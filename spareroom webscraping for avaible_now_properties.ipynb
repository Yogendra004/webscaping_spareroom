```python
#importing required libraries
import pandas as pd
import numpy as np
import requests
#beautiful soup library
from bs4 import BeautifulSoup
from termcolor import colored
import time
```

```python
#we are iterating over sveral pages to get all the result since every link has offset of 10 steps
pages=np.arange(0,30,10)
#creating a dataframe with 6 important columns
df=pd.DataFrame(columns=['Title','No_of_rooms','Location','Description','Price','Availibility'])
#for loop for pages
for page in pages:
    #storing url of website under URL variable
    URL='https://www.spareroom.co.uk/flatshare/?offset='+str(page)+'&search_id=1121063061&sort_by=age&mode=list'
    #requesting web to respond to get query for the required url
    data=requests.get(URL)
    #creating beautiful soup object and storing website content into it
    soup=BeautifulSoup(data.text,'html5lib')
    #spareroom has several lists with class "listing-result" under 1 unordered list
    #every list is itself an individual advertisment comprised of all information such as price, location, title etc
    spare_room=soup.find_all('li',class_="listing-result")


    # now we are pulling required informations from tag "li " by iterating over all the lists
    for sr in spare_room:
        #"a " is tag consists of availibilty information under class "advertDescription"
        # extracting only text out of it by".text"
        avail=sr.find('a',class_="advertDescription").strong.text
        
        # show only properties which are available now to shift
        if 'Now' in avail:
            #title information under "a" tag 
            title=sr.find('a')['title']
            #no of rooms under "em" tag with class "shortDescription"
            no_rooms=sr.find('em',class_="shortDescription").text[:6].replace("1","single")
            #location information under "span" tag under "listingLocation" class
            location=sr.find('span',class_="listingLocation").text
            #price information under "strong" tag under "listingPrice" class
            price=sr.find('strong',class_="listingPrice").text 
            #description information under "p" tag under "description" class
            desc=sr.find('p',class_="description").text.replace("\n","")
        
            avail=sr.find('a',class_="advertDescription").strong.text
            #passing extracted information into created dataframe 
            df=df.append({'Title':title,'No_of_rooms':no_rooms,'Location':location,'Description':desc.strip(),'Price':price,
                          'Availibility':avail},ignore_index=True)

        #print results
        print(colored(f"* {title}",'blue',attrs=['bold'])) 
        print(f""" \n{no_rooms} room  \nlocation:{location} \ndescription:{desc.strip()} \nprice:{price}\n{avail} \n 
            """)

    
    
    
```
